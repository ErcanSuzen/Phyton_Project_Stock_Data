### Extracting Data 

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
```

```python
Bitcoin=yf.Ticker("BTC-USD")
Amazon=yf.Ticker("AMZN")
Ford=yf.Ticker("F")
```

```python
Bitcoin.info
```

```python
Bitcoin_data = Bitcoin.history(period="1y")
Amazon_data=Amazon.history(period="1y")
Ford_data=Ford.history(period="1y")
```

```python
Bitcoin_data.reset_index(inplace=True)
Amazon_data.reset_index(inplace=True)
Ford_data.reset_index(inplace=True)
```

```python
Bitcoin_data.info()
```

```python
Bitcoin_data.isnull().sum()
```

```python
Bitcoin_data
```

```python
Bitcoin_data.plot(x="Date" , y="Close",ylabel="Bitcoin price",title="Bitcoin Stock - 1year", figsize=(10,10))
Amazon_data.plot(x="Date", y="Close", ylabel="Amazon price", title="Amazon Stock - 1year",figsize=(10,10))
Ford_data.plot(x="Date", y="Close", ylabel="Ford price", title="Ford Stock -1year",figsize=(10,10))
```
